## Make Essential Gene Predictions

```python
essential=[]
nonessential=[]
essDF= pd.DataFrame(columns=['CD','Ess?'])

y=0
for g in CDmod.genes:
    delete_model_genes(CDmod,[g])
    CDmod.optimize()
    growth=CDmod.optimize().f
    if growth<.0001:
        essential.append(g)
        essDF.loc[y] = pd.Series({'CD':g.id, 'Ess?':1})
    else:
        nonessential.append(g)
        essDF.loc[y] = pd.Series({'CD':g.id, 'Ess?':0})
    #print (g, CDmod.optimize().f)
    y=y+1
    undelete_model_genes(CDmod)
```

## Compare to R20291 Dataset

```python
dic = pickle.load(open('model_against_R20291_RERUN.p','rb'))
df = pd.DataFrame(columns=['R27','630'])
y=0
for k,v in dic.items():
    df.loc[y] = pd.Series({'630':v, 'R27':k})
    y=y+1
pd.set_option('display.max_rows', None)  
df
```

```python
nonModelGenes=0
listR27=[]
list630=[]
for x, row in df.iterrows():
    currentGeneR27=df.iloc[x]['R27']
    currentGene630=df.iloc[x]['630']
    try:
        CDmod.genes.get_by_id(currentGene630)
        list630.append(currentGene630)
        listR27.append(currentGeneR27)
    except:
        nonModelGenes=nonModelGenes+1
```

```python
essentiality = pd.DataFrame({'630': list630,'R27': listR27})
essentiality
```

```python
R27Ess=pd.read_csv('R27Ess.csv')
listR27Ess=[]
for g, row in R27Ess.iterrows():
    listR27Ess.append(R27Ess.iloc[g]['Essential Genes'])
```

```python
essentialIDs=[]
for g in essential:
    essentialIDs.append(g.id)
```

```python
listEssentiality=[]
listR27Essentiality=[]
y=0
for x, row in essentiality.iterrows():
    currentG=essentiality.iloc[x]['630']
    currentR27=essentiality.iloc[x]['R27']
    if currentR27 in listR27Ess:
        listR27Essentiality.append('Essential')
    else:
         if type(currentR27) != str:
                listR27Essentiality.append('No Homolog')
         else:
            listR27Essentiality.append('Non Essential')
    if currentG in essentialIDs:
        listEssentiality.append('Essential')
        y=y+1
    else:
        listEssentiality.append('Non Essential')
```

```python
essentiality2 = pd.DataFrame({'630': list630,'R27': listR27, 'Ess 630': listEssentiality, 'Ess R27': listR27Essentiality})
essentiality2
```

```python
Score=[]
for x, row in essentiality2.iterrows():
    ess630=essentiality2.iloc[x]['Ess 630']
    essR27=essentiality2.iloc[x]['Ess R27']
    
    if ess630 == essR27:
        if essR27=='Non Essential':
            Score.append('TP')
        else:
            Score.append('TN')     
    else:
        if essR27=='Essential':
            Score.append('FP')
        else:
            if essR27=='Non Essential':
                Score.append('FN')
            else:
                Score.append('No Hom')
```

```python
essentiality3 = pd.DataFrame({'630': list630,'R27': listR27, 'Ess 630': listEssentiality,
                              'Ess R27': listR27Essentiality, 'Score': Score})
essentiality3
```
